```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("Data.csv")
data.head()
```

```python
data.info()
```

```python

data['Date'].unique()
```

```python
lastest_date = '3/3/2020'
data[data['Date']==lastest_date]
```

```python
unique_countries = data[data.Latitude>0]['Location ISO Code'].unique()
unique_countries.sort
unique_countries
```

```python
country_confirmed_case = []
for i in unique_countries:
  country_confirmed_case.append(data[data.Latitude>0][data['Location ISO Code']==i][data['Date']==lastest_date].Latitude.sum())
     
```

```python
print(set(zip(unique_countries, country_confirmed_case)))
```

```python
unique_provinces = data['Province'][data.Latitude>1].unique()
unique_provinces
```

```python
province_confirmed_case = []
for i in unique_provinces:
  province_confirmed_case.append(data[data.Latitude>0][data['Province']==i][data['Date']==lastest_date].Latitude.sum())
```

```python
print(set(zip(unique_provinces, province_confirmed_case)))
```

```python
nan_indices = []
for i in range(len(unique_provinces)):
  if type(unique_provinces[i])==float:
    nan_indices.append(i)
```

```python
unique_provinces = list(unique_provinces)
province_confirmed_case = list(province_confirmed_case)
```

```python
for i in nan_indices:
  unique_provinces.pop(i)
  province_confirmed_case.pop(i)

unique_provinces
```

```python
plt.barh(unique_countries, country_confirmed_case)
plt.show()
```
